```python
# 1. Change all columns to object type
for col in df.columns:
    df[col] = df[col].astype(str)

# 2. Define lists of float and int columns
float_cols = ['loan_amount', 'loan_to_value_ratio', 'interest_rate',
              'rate_spread', 'total_loan_costs', 'total_points_and_fees',
              'origination_charges', 'discount_points', 'lender_credits',
              'property_value', 'income', 'tract_minority_population_percent',
              'ffiec_msa_md_median_family_income', 'tract_to_msa_income_percentage']

int_cols = ['loan_term', 'prepayment_penalty_term', 'intro_rate_period',
            'multifamily_affordable_units', 'tract_population', 'tract_owner_occupied_units',
            'tract_one_to_four_family_homes', 'tract_median_age_of_housing_units']

# 3. Convert float columns to float type
for col in float_cols:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)
    except (ValueError, TypeError):
        print(f"Error converting column {col} to float. Check for non-numeric values.")

# 4. Convert int columns to int type
for col in int_cols:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype(int)
    except (ValueError, TypeError):
        print(f"Error converting column {col} to int. Check for non-numeric values.")
```
